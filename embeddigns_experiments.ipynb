```python
import networkx as nx
from linear_pa import ScaleFreeDiGraph
from embeddings import PageRankClustering, PageRankInDegree
from visualisers import visualise_communities
from detectors import GPDCCommunities

import matplotlib.pyplot as plt
```

```python
init_graph = nx.MultiDiGraph()
init_graph.add_node(0)
graph = ScaleFreeDiGraph(init_graph=init_graph)
graph.grow(10000, alpha=0.4, beta=0.2, gamma=0.4, delta_in=1, delta_out=1)
```

```python
max_node = len(graph.get_graph())
```

```python
max_node
```

```python
embeddings = PageRankInDegree(graph.get_graph())
```

```python
data = embeddings.to_numpy()

plt.figure(figsize=(15, 7))

plt.xlabel('PageRank')
plt.ylabel('Clustering coefficient')
plt.scatter(x=data[:, 0], y=data[:, 1], alpha=0.3)

# plt.xscale('log')
# plt.yscale('log')

plt.show()
```

```python
graph.grow(1000, alpha=0.4, beta=0.2, gamma=0.4, delta_in=1, delta_out=1)
```

```python
len(graph.get_graph()) - max_node
```

```python
embeddings.update(graph.get_graph())
data = embeddings.to_numpy()

plt.figure(figsize=(15, 7))

plt.xlabel('PageRank')
plt.ylabel('Clustering coefficient')
plt.scatter(x=data[:max_node:, 0], y=data[:max_node, 1], alpha=0.3, color='blue')
plt.scatter(x=data[max_node:, 0], y=data[max_node:, 1], alpha=0.3, color='red')

# plt.xscale('log')
# plt.yscale('log')

plt.show()
```
